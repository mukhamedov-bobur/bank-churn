```python
# This script:
# 1. Imports required modules
# 2. Loads raw dataset
# 3. Performs K-fold cross validation and Selects SHAP-important features inside each fold
# 4. Calculates CV metrics using important features
# 5. Aggregate stable SHAP features
# 5. Optimizes classification threshold (Recall-focused)
# 6. Confusion matrix and ROC curve
# 7. Trains final production mode

# ===============================
# 1. Imports
# ===============================
import pandas as pd
import numpy as np
import shap
import json
from pathlib import Path

from xgboost import XGBClassifier

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (
    roc_auc_score,
    average_precision_score,
    precision_recall_curve,
    recall_score,
    f1_score,
    confusion_matrix,
    roc_curve
)
from sklearn.calibration import CalibratedClassifierCV

import matplotlib.pyplot as plt
from collections import Counter


# ===============================
# 2. Load dataset
# ===============================
df = pd.read_csv("TZ.csv")

TARGET = "—É—à–µ–ª_–∏–∑_–±–∞–Ω–∫–∞"
DROP_COLS = ["ID", "ID_–∫–ª–∏–µ–Ω—Ç–∞", "—Ñ–∞–º–∏–ª–∏—è"]

X = df.drop(columns=DROP_COLS + [TARGET])
y = df[TARGET].astype(int)

categorical_features = [
    "–≥–æ—Ä–æ–¥",
    "–ø–æ–ª",
    "–∞–∫—Ç–∏–≤–Ω—ã–π_–∫–ª–∏–µ–Ω—Ç",
    "–µ—Å—Ç—å_–∫—Ä–µ–¥–∏—Ç–∫–∞"
]

numerical_features = [
    c for c in X.columns if c not in categorical_features
]


# ===============================
# 3. CV setup with SHAP feature selection
# ===============================
N_SPLITS = 10
SHAP_CUMSUM = 0.95
MIN_FOLD_OCCURRENCE = 3
MAX_FEATURES = 25

kf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=42)

oof_probs = np.zeros(len(y))
oof_true = np.zeros(len(y))

selected_features_per_fold = []
scores_roc_auc = []
scores_pr_auc = []

def target_encode_train(X, y, categorical_cols, min_samples=10, smoothing=5):
    X_enc = X.copy()
    global_mean = y.mean()

    encodings = {}

    for col in categorical_cols:
        stats = (
            pd.concat([X[col], y], axis=1)
            .groupby(col)[y.name]
            .agg(["mean", "count"])
        )

        smooth = (
            (stats["count"] * stats["mean"] + smoothing * global_mean) /
            (stats["count"] + smoothing)
        )

        encodings[col] = smooth
        X_enc[col] = X[col].map(smooth).fillna(global_mean)

    return X_enc, encodings

def target_encode_apply(X, encodings, global_mean):
    X_enc = X.copy()

    for col, mapping in encodings.items():
        X_enc[col] = X[col].map(mapping).fillna(global_mean)

    return X_enc


for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
    print(f"\nüîπ Fold {fold + 1}")

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    # ===============================
    # Target Encoding (CV-safe)
    # ===============================

    X_train_enc, encodings = target_encode_train(
        X_train,
        y_train,
        categorical_features,
        smoothing=5
    )

    global_mean = y_train.mean()

    X_val_enc = target_encode_apply(
        X_val,
        encodings,
        global_mean
    )

    pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

    # ===============================
    # Base model (ALL features)
    # ===============================
    base_model = XGBClassifier(
        n_estimators=400,
        max_depth=4,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        min_child_weight=20,
        gamma=1,
        objective="binary:logistic",
        eval_metric="auc",
        scale_pos_weight=pos_weight,
        random_state=42,
        tree_method="hist"
    )

    base_model.fit(X_train_enc, y_train)

    # ===============================
    # SHAP on TRAIN only
    # ===============================
    explainer = shap.TreeExplainer(base_model)
    shap_values = explainer(X_train_enc)

    mean_abs_shap = np.abs(shap_values.values).mean(axis=0)

    shap_df = pd.DataFrame({
        "feature": X_train_enc.columns,
        "importance": mean_abs_shap
    })

    # ---- Aggregate SHAP by original feature
    shap_df["base_feature"] = shap_df["feature"]
    shap_df = (
        shap_df.groupby("base_feature", as_index=False)["importance"]
        .sum()
        .sort_values("importance", ascending=False)
        .reset_index(drop=True)
    )

    shap_df["cumsum_ratio"] = (
        shap_df["importance"].cumsum() / shap_df["importance"].sum()
    )

    fold_features = shap_df.loc[
        shap_df["cumsum_ratio"] <= SHAP_CUMSUM, "base_feature"
    ].head(MAX_FEATURES).tolist()

    if len(fold_features) < 5:
        fold_features = shap_df["base_feature"].head(5).tolist()

    selected_features_per_fold.append(fold_features)

    print(f"Selected top features for this fold: {len(fold_features)}")
    print(fold_features)

    # ===============================
    # Reduced datasets
    # ===============================
    X_train_sel = X_train_enc[fold_features]
    X_val_sel = X_val_enc[fold_features]

    # ===============================
    # Final model + calibration
    # ===============================
    final_model = XGBClassifier(
        n_estimators=600,
        max_depth=4,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        min_child_weight=20,
        gamma=1,
        objective="binary:logistic",
        eval_metric="auc",
        scale_pos_weight=pos_weight,
        random_state=42,
        tree_method="hist"
    )

    calibrated_model = CalibratedClassifierCV(
        estimator=final_model,
        method="isotonic",
        cv=3
    )

    calibrated_model.fit(X_train_sel, y_train)

    fold_probs = calibrated_model.predict_proba(X_val_sel)[:, 1]

    oof_probs[val_idx] = fold_probs
    oof_true[val_idx] = y_val
    scores_roc_auc.append(round(float(roc_auc_score(y_val, fold_probs)), 4))
    scores_pr_auc.append(round(float(average_precision_score(y_val, fold_probs)), 4))


# ===============================
# 4. CV metrics
# ===============================
print("\nROC-AUC per fold:", scores_roc_auc)
print(f"Mean ROC-AUC: {np.mean(scores_roc_auc):.4f} ¬± {np.std(scores_roc_auc):.4f}")

print("\nPR-AUC per fold:", scores_pr_auc)
print(f"Mean PR-AUC: {np.mean(scores_pr_auc):.4f} ¬± {np.std(scores_pr_auc):.4f}")


# ===============================
# 5. Aggregate stable SHAP features
# ===============================
feature_counter = Counter(
    f for fold_feats in selected_features_per_fold for f in fold_feats
)

stable_features = [
    f for f, c in feature_counter.items()
    if c >= MIN_FOLD_OCCURRENCE
]

print("\nStable SHAP-selected features:")
print(stable_features)
print(f"Total stable features: {len(stable_features)}")


# ===============================
# 6. Threshold optimization (Recall-focused)
# ===============================
MIN_PRECISION = 0.45

precision, recall, thresholds = precision_recall_curve(oof_true, oof_probs)
valid = precision[:-1] >= MIN_PRECISION

valid_idx = np.where(valid)[0]
best_idx = valid_idx[np.argmax(recall[valid_idx])]
best_threshold = thresholds[best_idx]

y_pred = (oof_probs >= best_threshold).astype(int)

print(f"\nOptimal threshold: {best_threshold:.3f}")
print(f"Recall: {recall_score(oof_true, y_pred):.3f}")
print(f"F1-score: {f1_score(oof_true, y_pred):.3f}")

cm = confusion_matrix(oof_true, y_pred)
print("\nConfusion Matrix:")
print(cm)


# ===============================
# 7. ROC Curve
# ===============================
fpr, tpr, _ = roc_curve(oof_true, oof_probs)
roc_auc = roc_auc_score(oof_true, oof_probs)

plt.figure(figsize=(7, 6))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0, 1], [0, 1], "--", color="gray")
plt.legend()
plt.grid(alpha=0.3)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (OOF predictions)")
plt.savefig('ROC-curve_xgboost')
plt.show()


# ===============================
# 8. Train final production model
# ===============================

# === Final target encoding (FULL DATA)

#stable_cat_features = [f for f in categorical_features if f in stable_features]
X_full_enc, final_encodings = target_encode_train(
    X,
    y,
    stable_features,
    smoothing=5
)

global_mean = y.mean()

# Keep only stable features
X_full_enc = X_full_enc[stable_features]

final_model = XGBClassifier(
    n_estimators=700,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=20,
    gamma=1,
    objective="binary:logistic",
    eval_metric="auc",
    scale_pos_weight=(y == 0).sum() / (y == 1).sum(),
    random_state=42,
    tree_method="hist"
)

final_model.fit(X_full_enc, y)


# ===============================
# 9. Save artifacts
# ===============================
ARTIFACT_DIR = Path("artifacts")
ARTIFACT_DIR.mkdir(exist_ok=True)

final_model.save_model(ARTIFACT_DIR / "final_model_xgb_te.json")

with open(ARTIFACT_DIR / "stable_features.json", "w", encoding="utf-8") as f:
    json.dump(stable_features, f, ensure_ascii=False, indent=2)

with open(ARTIFACT_DIR / "best_threshold.json", "w") as f:
    json.dump(
        {
            "threshold": float(best_threshold),
            "min_precision": MIN_PRECISION,
            "objective": "maximize_recall_given_precision_constraint"
savefig        },
        f,
        indent=2
    )

ARTIFACT_DIR = Path("artifacts")
ARTIFACT_DIR.mkdir(exist_ok=True)

with open(ARTIFACT_DIR / "target_encodings.json", "w", encoding="utf-8") as f:
    json.dump(
        {k: v.to_dict() for k, v in final_encodings.items()},
        f,
        ensure_ascii=False,
        indent=2
    )

with open(ARTIFACT_DIR / "global_mean.json", "w") as f:
    json.dump(float(global_mean), f)
```
